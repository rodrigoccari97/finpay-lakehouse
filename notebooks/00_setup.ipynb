```python
# ============================================
# 00_setup.ipynb
# Proyecto: FinPay Lakehouse
# ============================================

# ============================================
# VARIABLES BASE
# ============================================

CATALOG_NAME = "adbanalitycs"

BRONZE_SCHEMA = "bronze"
SILVER_SCHEMA = "silver"
GOLD_SCHEMA   = "gold"

VOLUME_SCHEMA = "default"
VOLUME_NAME   = "vol_landing"

# ============================================
# CREAR SCHEMAS
# ============================================

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {CATALOG_NAME}.{BRONZE_SCHEMA}
""")

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {CATALOG_NAME}.{SILVER_SCHEMA}
""")

spark.sql(f"""
CREATE SCHEMA IF NOT EXISTS {CATALOG_NAME}.{GOLD_SCHEMA}
""")

print("Schemas creados correctamente")

# ============================================
# CREAR VOLUME
# ============================================

spark.sql(f"""
CREATE VOLUME IF NOT EXISTS
{CATALOG_NAME}.{VOLUME_SCHEMA}.{VOLUME_NAME}
""")

print("Volume creado correctamente")

# ============================================
# RUTA BASE DEL VOLUME
# ============================================

BASE_VOLUME_PATH = f"/Volumes/{CATALOG_NAME}/{VOLUME_SCHEMA}/{VOLUME_NAME}"

# ============================================
# CREAR DIRECTORIOS PRINCIPALES
# ============================================

directories = [
    "transactions",
    "merchants",
    "users",
    "metadata",
    "metadata/checkpoints",
    "metadata/checkpoints/transactions",
    "metadata/checkpoints/merchants",
    "metadata/checkpoints/users",
    "metadata/schema",
    "metadata/schema/transactions",
    "metadata/schema/merchants",
    "metadata/schema/users"
]

# ============================================
# CREAR CARPETAS
# ============================================

for directory in directories:

    path = f"{BASE_VOLUME_PATH}/{directory}"

    dbutils.fs.mkdirs(path)

    print(f"Directorio creado: {path}")

# ============================================
# VALIDACION FINAL
# ============================================

print("\n===================================")
print("SETUP FINALIZADO CORRECTAMENTE")
print("===================================")

print(f"\nCatalogo: {CATALOG_NAME}")

print(f"""
Schemas:
- {BRONZE_SCHEMA}
- {SILVER_SCHEMA}
- {GOLD_SCHEMA}
""")

print(f"Volume:")
print(f"- {VOLUME_NAME}")

print(f"\nRuta base:")
print(BASE_VOLUME_PATH)
```
